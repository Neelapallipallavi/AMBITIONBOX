```python
company=[]
ratings=[]
review=[]
sector=[]
location=[]
counts=[]
years=[]
employees=[]
salaries=[]
l=list(range(1,15))
for i in l:
    url = ('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page=1')
    page=requests.get(url,headers={'User-Agent':'Mozilla/5.0'})
    soup=BeautifulSoup(page.text)
    
    
    for i in soup.find_all('h2',class_="company-name bold-title-l"):
        company.append(i.text.split()[0])
        
    
    for i in soup.find_all('div',class_="rating-wrapper"):
        ratings.append(i.text.split()[0])
        
    
    for i in soup.find_all('a',class_="review-count sbold-Labels"):
        review.append(i.text.split()[0].split('(')[1])
        
    
    
    for i in soup.find_all('div',class_='company-basic-info'):
        sector.append(i.text.split()[0])
        
    cit=soup.find_all('div',class_='company-basic-info')
    
    
    for i in cit:
        a=i.text
        b=re.findall(r'(\w+),',a)  
        if b:
            location.append(b[0])
        else:
            location.append(np.nan)
        
    data=soup.find_all('div',class_='company-basic-info')
    
    
    for i in data:
        a=i.text
        b=re.findall(r'\d+\smore',a)
        counts.append(b[0])
        
    y=soup.find_all('div',class_='company-basic-info')
    
    
    for i in y:
        a=i.text
        b=re.findall(r'\d+\syears',a)
        years.append(b[0])
        
    e=soup.find_all('div',class_='company-basic-info')
    
    
    for i in e:
        a=i.text
        b=re.findall(r'\d \w+\+ |\d+\w-\d |\d+\w+\-\d+\w',a)
        if b:
            employees.append(b[0])
    
    for i in soup.find_all('ul',class_='company-action-center'):
        salaries.append(i.text.split()[4])
        
dictionary={
        "Company":company,
        "Ratings":ratings,
        "Review":review,
        "Sector":sector,
        "Location":location,
        "Counts":counts,
        "Years":years,
        "Employees":employees,
        "Salaries":salaries
}   
df1=pd.DataFrame(dictionary)
df1
```
